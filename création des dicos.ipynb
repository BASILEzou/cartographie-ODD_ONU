```python
import pandas as pd
from tqdm import tqdm
import pickle
##import plotly.express as px

# Charger les données départementales
données_dep = pd.read_csv("C:\\Users\\basil\\OneDrive\\Documents\\python\\perso\\datanalys\\ODD_CSV(1)\\ODD_DEP.csv", encoding='latin1', sep=';')
#la fiche info de l'INSEE
fiche_info = pd.read_excel("C:\\Users\\basil\\OneDrive\\Documents\\python\\perso\\datanalys\\dico.xlsx")
#charger les données régionales
données_reg = pd.read_csv("C:\\Users\\basil\\OneDrive\\Documents\\python\\perso\\datanalys\\ODD_CSV(1)\\ODD_REG.csv", encoding='latin1', sep=';')
```

```python
#dictionnaire qui relie chaque var (variable total =une variable et un sous-champ) à : variable, sous-champ, indicateur, libellé, ODD, unité, libellé du sous-champs
fiche_info['var'] = fiche_info['Variable'] + ', ' + fiche_info['Sous-champ']
dico_variables = {
    var : {"variable": Variable, "Sous-champ": sous_champ, "indicateur": indicateur, "libellé": libellé, "ODD": ODD, "Unité": unité, "libellé_sous" : libellé_sous }
    for var, Variable, sous_champ, indicateur, libellé, ODD, unité, libellé_sous in zip(fiche_info['var'], fiche_info['Variable'], fiche_info['Sous-champ'], fiche_info['Numero indicateur'], fiche_info['Libellé de la variable'], fiche_info['ODD1'], fiche_info['Unité'], fiche_info['Libellé du sous-champ'])
}

with open("dico_variables.pkl", "wb") as f:
    pickle.dump(dico_variables, f)
```

```python
années = données_dep.columns[10:].astype(str).tolist()
années_reg = données_reg.columns[10:].astype(str).tolist()
départements = list(set(données_dep['libgeo']))
régions = list(set(données_reg['libgeo']))
var = données_dep['variable'] + ', ' + données_dep['sous_champ']

#fonction pour créer un dictionnaire qui relie une année d'un département à un dictionnaire avec toutes les sous_champs
def dico_an_dep(an, dep):
    dico={}
    df_dep = données_dep[données_dep['libgeo'] == dep]
    for i, row in df_dep.iterrows():
        cle = str(row['variable']) + ', ' + str(row['sous_champ'])
        valeur = row[an]
        dico[cle] = valeur
    return dico

#fonction pour créer un dictionnaire qui relie 1 département à un dictionnaire des années 
def dico_dep(dep):
    dico={}
    for an in années:
        dico[an] = dico_an_dep(an, dep)
    return dico

#fonction pour créer un dictionnaire qui relie une année d'un département à un dictionnaire avec toutes les sous_champs
def dico_an_reg(an, reg):
    dico={}
    df_reg = données_reg[données_reg['libgeo'] == reg]
    for i, row in df_reg.iterrows():
        cle = str(row['variable']) + ', ' + str(row['sous_champ'])
        valeur = row[an]
        dico[cle] = valeur
    return dico

#fonction pour créer un dictionnaire qui relie 1 département à un dictionnaire des années 
def dico_reg(reg):
    dico={}
    for an in années_reg:
        dico[an] = dico_an_reg(an, reg)
    return dico



```

```python
#création du dictionnaire avec toutes les données pour les reg

mega_dico_reg = {}
for reg in tqdm(régions):
    mega_dico_reg[reg] = dico_reg(reg)

with open("mega_dico_reg.pkl", "wb") as f:
    pickle.dump(mega_dico_reg, f)
```

```python
#création du dictionnaire avec toutes les données pour les dep

mega_dico = {}
for dep in tqdm(départements):
    mega_dico[dep] = dico_dep(dep)

with open("mega_dico.pkl", "wb") as f:
    pickle.dump(mega_dico, f)
```
